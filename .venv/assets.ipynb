```python
import pandas as pd
import requests
from dagster import asset, Output, AssetMaterialization

@asset(group_name="ingestion")
def raw_coin_data():
    """Fetch real-time data from CoinGecko."""
    url = "https://api.coingecko.com/api/v3/simple/price?ids=bitcoin,ethereum&vs_currencies=eur"
    data = requests.get(url).json()
    
    df = pd.DataFrame([
        {"ticker": "BTC", "price": data["bitcoin"]["eur"]},
        {"ticker": "ETH", "price": data["ethereum"]["eur"]}
    ])
    
    # Save locally to mimic a landing zone
    df.to_parquet("data/raw_coin_data.parquet")
    return df
```
